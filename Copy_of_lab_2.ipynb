# phase 2
# Data Cleaning

```python
clean = data.filter(lambda row: "" not in row)

print(clean.take(5))
```

```python
print(type(header))
print(header)
```

```python
for i in clean.take(1)[0]:
    print(f"{i} --  {type(i)} -- {clean.take(1)[0].index(i)}")
```

```python
from datetime import datetime

def safe_int_conversion(value):
    try:
        return int(value.strip())
    except ValueError:
        return None

def safe_datetime_conversion(date_string):
    try:
        return datetime.strptime(date_string, '%m/%d/%Y %I:%M:%S %p')
    except ValueError:
        return None

numeric_clean = clean.map(lambda row: (row[0],
                                       row[1],
                                       safe_datetime_conversion(row[2]), # Apply safe conversion here
                                       tuple(row[3:10]),
                                       safe_int_conversion(row[10]),
                                       safe_int_conversion(row[11]),
                                       safe_int_conversion(row[12]),
                                       safe_int_conversion(row[13]),
                                       safe_int_conversion(row[14]),
                                       safe_int_conversion(row[15]),
                                       safe_int_conversion(row[16]),
                                       safe_int_conversion(row[17]),
                                       safe_datetime_conversion(row[18]), # Apply safe conversion here
                                       row[19],
                                       row[20],
                                       ))
```

# Phase 3
# Core RDD Transformations

```python
# Assuming 'numeric_clean' is an RDD of tuples as per its current definition
# The fourth element (index 3) is a list containing Block, IUCR, Primary Type, Description, etc.
# Primary Type is at index 2 within that list (row[3][2])
# Description is at index 3 within that list (row[3][3])

words_rdd = numeric_clean.flatMap(
    lambda row: (row[3][2].lower().split(" ") if row[3][2] else []) + \
                (row[3][3].lower().split(" ") if row[3][3] else [])
).filter(lambda word: word != '') # Filter out empty strings that result from split

word_counts = words_rdd.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

# Display the top 10 most frequent words
print("Top 10 most frequent words from Primary Type and Description:")
for word, count in word_counts.top(10, key=lambda x: x[1]):
    print(f"{word}: {count}")
```

# Phase 4
# Key-Value Operations


```python
numeric_clean.take(1)
```

```python
numeric_clean.reduceByKey(lambda a,b: a+b)
```

```python
numeric_clean_kv = numeric_clean.map(lambda row: (row[0], row))
result = numeric_clean_kv.groupByKey().take(5)

for key, values in result:
    print(f"Key: {key}")
    for value in values:  # Iterate through the iterable
        print(f"  → {value}")
    print("-" * 50)
```

```python

numeric_clean_kv = numeric_clean.map(lambda row: (row[3][2], row))
sorted_result = numeric_clean_kv.sortByKey().take(5)

for key, value in sorted_result:
    print(f"Key: {key}, Value: {value}")
```

```python

numeric_clean_kv = numeric_clean.map(lambda row: (row[3][2], row))
counts = numeric_clean_kv.countByKey()  # Returns a Python dict, NOT an RDD

# Display top 5 counts
top_5 = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:5]
for key, count in top_5:
    print(f"{key}: {count}")
```

```python

numeric_clean_kv = numeric_clean.map(lambda row: (row[3][2], 1))
reduced_result = numeric_clean_kv.reduceByKey(lambda a, b: a + b).take(5)

for key, count in reduced_result:
    print(f"{key}: {count}")
```
